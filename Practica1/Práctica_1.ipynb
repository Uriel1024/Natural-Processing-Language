```python
"""
el patron siempre busca en la primer letra una 'r' minuscula, luego busca cualquier
caracter que se repita n veces con .+, al final busca una 'g' minuscula.
"""
patron = r'(.*)r(.*)g(.*)'
for i in pruebas:
  if re.search(patron,i):
    print(f"La cadena {i} \t\t es valida.")
  else:
    continue
    print(f"La cadena {i} no es valida.")
```

**2** Contenga una cadena 'i' minúscula, seguida por otros caracteres, seguida por una 'd' mínuscula, los caracteres i y d pueden ser minúscula o mayuscula.

```python

"""
el patron siempre busca en la primer letra una 'i' minuscula o mayuscula, luego busca cualquier
caracter que se repita n veces con .+, al final busca una 'd' minuscula o mayuscula.
"""
patron = r'(.*)(I|i)(.*)(d|D)(.*)'
for i in pruebas:
  if re.search(patron,i):
    print(f"La cadena {i} \t\t es valida.")
  else:
    continue
    print(f"La cadena {i} no es valida.")
```

**3** Contengan en la descripción de gastos una "a" minúscula, seguida de otros caracteres, y al menos tres dígitos en cualquier orden; la letra "a" puede aparecer antes o despúes de los dígitos.  

```python
"""
el patron primero toma cualquier caracter n veces (.+), luego pregunta si hay una 'a',
despues busca al menos 3 digitos cualesquiera, despues sigue buscando si existe cualquier caracter,
despues un esta un or | que nos indica el otro caso, que primero tima cualquier caracter (.+), luego
busca al menos 3 digitos cualesquiera y una 'a', al final toma n caracteres cualesquiera (.+).
"""
patron = r'(.*)(\d{3,})(.*)a(.+)(\d{3,}(.*))'
for i in pruebas:
  if re.search(patron,i):
    print(f"La cadena {i} \t\t es valida.")
  else:
    continue
    print(f"La cadena {i} no es valida.")
```

**4** Describa comidas que cuesten al menos 100.00

```python
"""
Primero busca cualquier caracter, despues busca que los alimentos tengan al menos
3 digitos , luego corta el punto . , ya que en caso de que no se coloque y algun
alimento cueste menos de 100 pero tenga varios centavos lo tomara como que si este tuviera
ms de 3 digitos.
"""
patron = r'(.*)(\d{3,}\.(\d))'

for i in pruebas:
  if re.search(patron,i):
    print(f"El alimento {i} \t\t cuesta mas de 100.")
  else:
    continue
    print(f"El alimento {i} cuesta menos de 100.")
```

**5** Contenga un 'a', seguida por una 'c', seguida por una 'e'; pudiendo existir otros caracteres entre la 'a' y la 'c', entre la 'c' y la 'e', estos pueden ser minusculas o mayusculas, sin embargo entre la 'a' y la 'c' no puede estar la 'b' y entre la 'c' y la 'e' no puede estar la 'd'.

```python
"""
el patron busca si hay algo antes o no de la primer a, luego busca cualquier caracter
excepto la b, despues busca una c, despues cualquier caracter n- 0ve
"""
patron = r'(.*)a([^b]*)c([^d]*)e(.*)'

for i in pruebas:
  if re.search(patron,i):
    print(f"La cadena {i} \t\t es valida.")
  else:
    continue
    print(f"La cadena {i} no es valida.")
```

**6** Identificar títulos de películas producidas antes de 1995.

```python
peliculas = [
    'The Shawshank Redemption (1994)',
    "The Godfather (1972)",
    "The Godfather: Part II (1974)",
    "2001: A Space Odyssey (1968)",
    "The Good, the Bad and the Ugly (1966)",
    "12 Angry Men (1957)",
    "Schindler's List (1993)",
    "The Lord of the Rings: The Return of the King (2003)",
    "Fight Club (1999)",
    "2010: The Year We Make Contact (1984)",
    "101 Dalmatians (1996)"
]
"""
El patron primero ignora todos los caracteres antes del primer 19, despues busca
1900, 1901, 1902,...,1989 o  1991,1992,1993,1994
"""
patron = r'(.*)((1[8-9][0-8]\d)|(199[0-4]))'

for i in peliculas:
  if re.search(patron,i):
    print(f"La pelicula {i} se estreno antes de 1995.")
  else:
    print(f"La pelicula {i} se estreno despues de 1995.")
```

**9. Simplificar direcciones IPv6.** \

Una dirección IP está compuesta de 8 bloques de números hexadecimales. Cada
bloque está compuesto de cuatro dígitos hexadecimales y los bloques están
separados por el signo “:” (dos puntos). Adicionalmente, existen las siguientes
reglas que se pueden combinar para simplificar y reducir el tamaño de una
dirección IPv6:
a) Los ceros al principio de un bloque pueden ser omitidos.
b) Los bloques consecutivos compuestos únicamente de ceros pueden ser
pueden ser reemplazados por “:” (dos puntos). Esto solo se puede hacer
una vez.

```python
direccion = ['2001:0002:0000:0000:ff00:0db8:0000:0329', '2607:f0d0:1002:0051:0000:0000:0000:0004']
"""
El primer patron simplementa reemplaza los tres ceros por espacios vacios.
El segundo patron remplaza los tres puntos por puntos por solo dos.
"""
patron = r'0{3,}'
patron2 = r':{3,}'
for i in direccion:
  a = re.sub(patron,"",i)
  a = re.sub(patron2,"::",a)
  print(a)
```
